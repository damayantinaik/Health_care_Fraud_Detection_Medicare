**As there are some columns where it seems to have outliers, So let us replace the outliers with 90th percentiles.**

```python
#columns = ['InscClaimAmtReimbursed', 'InpatientClaimPeriod', 'DurationInHospital', 'IPAnnualReimbursementAmt', 'IPAnnualDeductibleAmt', 'OPAnnualReimbursementAmt', 'OPAnnualDeductibleAmt']
```

```python
def imputation_for_outliers(dataframe, cols, percentile):
    for col in cols:
        percentile_value = dataframe[col].quantile(q = percentile)
        #dataframe2 = dataframe.copy()
        dataframe.loc[dataframe[col]> percentile_value, col] = percentile_value
    return dataframe
```

```python
columns = X_train2.columns.tolist()
columns
```

```python
outliers_filled_X_train = imputation_for_outliers(X_train2, columns, 0.90)
```

```python
from sklearn.preprocessing import StandardScaler
def standard_scaling(df, columns):
    scale = StandardScaler()
    df[columns] = scale.fit_transform(df)
    return df
```

```python
cols = outliers_filled_X_train.columns.tolist()
scaled_X_train = standard_scaling(outliers_filled_X_train, cols)
```
