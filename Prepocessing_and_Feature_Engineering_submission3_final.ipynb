# Feature Engineering

Before carrying out the feature engineering, let us separate dataframe into features and target. 

```python
X = data.drop(columns = ['Provider', 'PotentialFraud'], axis = 1)
y = data['PotentialFraud']
```

```python
X.shape, y.shape
```

**Before carrying out the feature engineering, let us divide the X and y into train and test datasets.**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Reset the index column of X_train as it is random
X_train2 = X_train.reset_index()
X_train2.head()
```

```python
# Drop the 'index' column
X_train2.drop(columns = ['index'], inplace = True)
X_train2.head()
```

```python
# Assign the same index to y_train as X_train2.
y_train.index = X_train2.index
print(y_train.head())
print(y_train.shape)
```

```python
# Carry out the same above steps to replace the index_column of X_test and y_test. 
X_test2 = X_test.reset_index()
X_test2.drop(columns = ['index'], inplace = True)
y_test.index = X_test2.index
```

**Let us check the outliers in both train and test dataframe, and replace that with approprate value:**

```python
X_train2.describe(percentiles = [0.25, 0.50, 0.75, 0.90, 0.95, 0.99])
```

As there are some columns where it seems to have outliers, So let us Let us write a function to replace the outliers in all columns.

```python
def imputation_for_outliers(dataframe1, dataframe2,  cols, percentile):
    for col in cols:
        percentile_value = dataframe1[col].quantile(q = percentile)
        #dataframe2 = dataframe.copy()
        dataframe1.loc[dataframe1[col]> percentile_value, col] = percentile_value
        dataframe2.loc[dataframe2[col]> percentile_value, col] = percentile_value
    return dataframe1, dataframe2
```

Let us replace the outliers with 90th percentiles as it seems to be reasonable.

```python
# List of columns in X_train2 and X_test2
columns = X_train2.columns.tolist()
# Apply the imputation function both on X_train2 and X_test2
outliers_filled_X_train, outliers_filled_X_test  = imputation_for_outliers(X_train2,X_test2, columns, 0.90)
```

```python
outliers_filled_X_train.describe(percentiles = [0.25, 0.50, 0.75, 0.90, 0.95, 0.99])
```

```python
outliers_filled_X_test.describe(percentiles = [0.25, 0.50, 0.75, 0.90, 0.95, 0.99])
```

**Let us scale few numerical columns:**

**Write a function to standardscale the selected numerical data:**

```python
from sklearn.preprocessing import StandardScaler
def standard_scaling(df1, df2, columns):
    scaler = StandardScaler().fit(df1)
    df1[columns] = scaler.transform(df1)
    df2[columns] = scaler.transform(df2)
    return df1, df2
```

```python
cols = outliers_filled_X_train.columns.tolist()
scaled_X_train, scaled_X_test = standard_scaling(outliers_filled_X_train, outliers_filled_X_test, cols)
```

```python
scaled_X_train.head()
```

```python
scaled_X_train.shape, y_train.shape
```

```python
scaled_X_train.isnull().sum().sum()
```

```python
scaled_X_test.shape, y_test.shape
```

```python
scaled_X_test.isnull().sum().sum()
```

## Save the file for model building

```python
scaled_X_train.to_csv('X_train3.csv')
y_train.to_csv('y_train3.csv')
scaled_X_test.to_csv('X_test3.csv')
y_test.to_csv('y_test3.csv')
```
