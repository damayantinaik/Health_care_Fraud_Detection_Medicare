**As there are some columns where it seems to have outliers, So let us replace the outliers with 99th percentiles.**

```python
columns = ['InscClaimAmtReimbursed', 'InpatientClaimPeriod', 'DurationInHospital', 'IPAnnualReimbursementAmt', 'IPAnnualDeductibleAmt', 'OPAnnualReimbursementAmt', 'OPAnnualDeductibleAmt']
```

```python
def imputation_for_outliers(dataframe, cols, percentile):
    for col in cols:
        percentile_value = dataframe[col].quantile(q = percentile)
        dataframe.loc[dataframe[col]> percentile_value, col] = percentile_value
    return dataframe
```

```python
outliers_filled_X_train = imputation_for_outliers(X_train,columns, 0.99)
outliers_filled_X_train.head()
```

```python
outliers_filled_X_train[['InscClaimAmtReimbursed', 'InpatientClaimPeriod', 'DurationInHospital', 'IPAnnualReimbursementAmt', 'IPAnnualDeductibleAmt', 'OPAnnualReimbursementAmt', 'OPAnnualDeductibleAmt']].describe(percentiles = [0.25, 0.50, 0.75, 0.90, 0.95, 0.99])
```

**Let us scale few numerical columns:**

```python
cols_to_scaling = ['InscClaimAmtReimbursed', 'DeductibleAmtPaid', 'InpatientClaimPeriod', 'OutpatientClaimPeriod', 'DurationInHospital','NoOfMonths_PartACov', 'NoOfMonths_PartBCov', 'IPAnnualReimbursementAmt', 'IPAnnualDeductibleAmt', 'OPAnnualReimbursementAmt', 'OPAnnualDeductibleAmt']
```

**Write a function to standardscale the selected numerical data:**

```python
from sklearn.preprocessing import StandardScaler
def standard_scaling(df, cols):
    scale = StandardScaler()
    for column in cols:
        df[column] = scale.fit_transform(df[[column]])
    return df  
```

```python
scaled_X_train = standard_scaling(outliers_filled_X_train, cols_to_scaling)
```
